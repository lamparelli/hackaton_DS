```python
import numpy as np
import pandas as pd
import seaborn as sns
import os
```

Importo dati

```python
stb_data = pd.read_csv("data" + os.sep + "datiOutput.csv")
```

Rinomino colonne

```python
stb_data.rename(columns={stb_data.columns[0]: "ID", "_time": "time"}, inplace=True)
```

Elimino colonna ID

```python
stb_data.drop(columns="ID", inplace=True)
```

Rendo lowercase colonne

```python
stb_data.columns = stb_data.columns.str.lower()
```

Ordino alfabeticamente le colonne

```python
stb_data.sort_index(axis=1, inplace=True)
```

Elimino le colonne che non servono

```python
pd.set_option("max_rows", None)
seriesCols = pd.Series(stb_data.columns)
#tutte le colonne da eliminare in quanto hanno _split
deleteCol = seriesCols[(seriesCols.str.contains("_split") & (~seriesCols.str.contains("aamp_abr_bw_split") & (~seriesCols.str.contains("ap_info_split"))))] 
#elimino le colonne
stb_data.drop(columns=deleteCol, inplace=True)
```

Split delle colonne con più campi di intresse in più colonne

Guardo sample di N valori non vuoti per ogni colonna

```python
pd.set_option("max_columns", None)
pd.set_option("max_colwidth", None)
```

```python
# NB: NON mostra le prime N righe con nessuno valore NA.
# I dati in ogni colonna sono slegati dai dati nelle altre colonne.
# Mostra i primi N dati non-NA in ogni colonna.

numValoriSamplePerColonna = 10

vals = []
for col in stb_data.columns:
    sampleValoriColonna = stb_data[col].unique()[:numValoriSamplePerColonna]
    sampleValoriColonna = pd.Series(sampleValoriColonna).rename(col)
    vals.append(sampleValoriColonna)
sample = pd.concat(vals, axis=1)
sample
```
